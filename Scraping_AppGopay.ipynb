Scraping Ulasan Pengguna terhadap Aplikasi Gopay di Google Play Store

```python
!pip install google-play-scraper

from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
import numpy as np
```

```python
scrapreview = reviews_all(
    'com.gojek.gopay',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=11000
)
```

```python
import csv

with open('ulasan_gopay.csv', mode='w', newline='', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow(['Review'])
  for review in scrapreview:
    writer.writerow([review['content']])
```

Loading and Cleaning Dataset

```python
app_review_df = pd.DataFrame(scrapreview)
app_review_df.shape
```

```python
app_review_df.head()
```

```python
app_review_df = app_review_df[['reviewId', 'content']]
app_review_df.head()
```

```python
app_review_df.info()
```

```python
clean_df = app_review_df.drop_duplicates()
clean_df.duplicated().sum()
```

```python
clean_df = clean_df.dropna()
clean_df.isnull().sum()
```

```python
sample_df = clean_df.sample(n=50000, random_state=42)
```
